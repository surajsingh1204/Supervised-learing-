```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
# to plot the function
def plot_function(func):
    x_val = np.linspace(-10,10,100)
    y_val = [func(x) for x in x_val]
    #print("x values are :",x_val)
    #print("y values are :",y_val)
    plt.plot(x_val, y_val)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()

plot_function(lambda x: x**2)
```

```python
# to plot the derivative of the function
def plot_derivative(func):
    x_val = np.linspace(-10,10,100)
    delta_x = 0.0001
    y_val = (func(x_val + delta_x) - func(x_val)) / delta_x
    plt.plot(x_val, y_val)

plot_derivative(lambda x: x**2)
```

```python
# to find the gradient descent
def gradient_descent(func,w):
    list_of_wts = []
    weight = w
    learning_rate = 0.1
    delta = 0.0001
    for i in range(100):
        derivative = (func(weight + delta) - func(weight)) / delta
        weight = weight - derivative * learning_rate
        list_of_wts.append(weight)
    return list_of_wts

gradient_descent(lambda x: x**2 + 4*x +3 , 10)
# values oscillating if there is no learning rate
# after adding the learning rate it is stabilized
```

```python
def f(x):
    return x**2
x = np.linspace(-10,10,100)
dfdx = np.gradient(f(x),x)
plt.plot(x, f(x), label ='f(x)')
plt.plot(x, dfdx, label='dfdx')
plt.legend()
plt.show()
```

```python
def f(x):
    return np.sin(x)
x = np.linspace(-10,10,100)
dfdx = np.gradient(f(x),x)

fig, ax = plt.subplots()
ax.plot(x, f(x), label='f(x)')
ax.plot(x, dfdx, label='df/dx')
ax.legend()
plt.show()
```
