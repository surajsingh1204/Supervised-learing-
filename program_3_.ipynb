```python
#program -3write a python to calculate gradient descent of a machine learning model
def gradient_descent(func,w):
    list_of_weights = []
    weight = w
    delta = 0.0001
    for i in range(100):
        derivative = (func(weight+delta)-func(weight))/delta
        weight=weight-derivative*0.1
        list_of_weights.append(weight)
    return list_of_weights
gradient_descent(lambda x:x**2+4*x+3,10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x) # Indented this line to be part of the function body

x = np.linspace(-5, 5, 1000)
dfdx = np.gradient(f(x), x)
fig, ax = plt.subplots()
ax.plot(x, f(x), label='f(x)')
ax.plot(x, dfdx, label='df/dx')
ax.legend()
plt.show()
```
